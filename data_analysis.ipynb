```python
# get the orthodox churches
orthodox_church_m = ['ARAPO_M', 'GRKAD_M', 'BEOC_M','ACROC_M']
# create the ratio of orthodox churches per county
religion['orthodox_ratio'] = religion.loc[:,orthodox_church_m].sum(axis=1) / religion.TOTPOP *100
# identify the 15 counties with the highest orthodox church ratio
religion.nlargest(15, 'orthodox_ratio')
```

```python
# get all the different churches
c_cols=[col for col in religion.columns if '_M' in col]
total_ch = len(c_cols)
# identify the number of churches present in each county
tmp_df=religion.filter(regex='_C')
tmp_df.loc[:,'num_ch']=tmp_df.astype(bool).sum(axis=1)
# get the ratio of churches per county to total available religions in the USA
religion.loc[:,'num_ch'] = tmp_df.loc[:,'num_ch']
religion.loc[:,'num_ch_ratio']= religion.loc[:,'num_ch']/total_ch
# plot the 3 most extreme counties based on different churches
extreme_counties = religion.nlargest(3,'num_ch')
plt.hist(extreme_counties['CNAME'],weights=extreme_counties.num_ch)
plt.show()
# plot the 3 most extreme counties based on different churches as a ratio of total different religions
extreme_counties_ratio = religion.nlargest(3,'num_ch_ratio')
plt.hist(extreme_counties['CNAME'],weights=extreme_counties.num_ch_ratio)
plt.show()
```

```python
import numpy as np
import pandas as pd
# plotting packages
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
air = pd.read_csv('Airlines.csv')
air
```

```python
air.shape
```

```python
air.describe()
```

```python
# summary of data types and missing values
air.info()
```

```python
# plot histograms of features
air.hist(figsize=(10,10))
plt.show()
```

```python
# check if Dest column includes all the airports of origin and thus is exhaustive
assert len(pd.concat([air.Origin,air.Dest]).unique()) == len(air.Dest.unique())
air.Origin.isin(air.Dest).all()
```

```python
# get the average delay of each airline
air.groupby(by=['UniqueCarrier']).mean().sort_values(by='ArrDelay',ascending=False)
```

```python
# get the average delay of airport
air.groupby(by=['Dest']).mean().sort_values(by='ArrDelay',ascending=False)
```

```python
# get the average delay of each airline in each airport
avg_delays = air.groupby(by=['Dest','UniqueCarrier']).mean()
```

```python
avg_delays.sort_values(by='ArrDelay',ascending=False)
```

```python
df = pd.get_dummies(air)
```

```python
X = df.drop(columns='ArrDelay')
y = df.ArrDelay
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer
# model selection
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
# models
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
# metrics
from sklearn.metrics import mean_squared_error as MSE
from sklearn.metrics import r2_score
```

```python
random = 123
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.2, random_state=random)
```

```python
# model creation
models = list()
models.append( ('Linear',LinearRegression()) )
models.append( ('Lasso',Lasso()) )
models.append( ('Ridge',Ridge()) )
models.append( ('Elastic',ElasticNet()) )
models.append( ('DTRegressor',DecisionTreeRegressor(random_state=random)) )
#models.append( ('GBR',GradientBoostingRegressor(random_state=random)) )
models.append( ('KNNR',KNeighborsRegressor()) )
models.append( ('SVR',SVR()) )
```

```python
# Use cross validation on the training data on all models
model_results = dict()
for model_name, model in models:
    #kfold = KFold(n_splits=10, shuffle=True, random_state=random)
    cv_score = cross_val_score(model, X, y , cv=10, scoring='neg_mean_squared_error')
    model_results[model_name] = cv_score
    
    # print RMSE, mean and std of cross val score for each model
    print(f'{model_name} -> rmse:{np.sqrt(-cv_score).mean():.2f}, mean:{cv_score.mean():.2f} , std:+/-{cv_score.std():.2f}')
    # print R^2 score for each model
    y_pre = model.fit(X_train, y_train).predict(X_test)
    print(f'R^2: {r2_score(y_test,y_pre):.4f}\n')
    
# using boxplots to compare model performance
fig = px.box(data_frame=model_results,
             orientation = 'v',
             title='Box plot of cross validation scores for each model'
             )
fig.update_layout({'xaxis':{'title':{'text':'Regression Models'}},
                   'yaxis':{'title':{'text':'Negative MSE'}}
                  })
fig.show()
```
